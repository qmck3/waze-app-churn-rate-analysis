### Handling missing/null values

Compare the summary statistics of the 700 rows that are missing labels with summary statistics of the rows that are not missing any values.

**Question:** Is there a discernible difference between the two populations?


```python
# Isolate rows without null values
not_null_df = df[~df['label'].isnull()]
# Display summary stats of rows without null values
not_null_df.describe()
```

```python
# Isolate rows with null values
null_rows = df[df['label'].isnull()]
# Display summary stats of rows with null values
null_rows.describe()
```

Comparing the mean, standard deviation and median between the missing label rows and the rows without missing values, we can see they are not significantly different. And since these values are categorical and they only contains about 5% of the data, I choose to delete these rows.

```python
df = df.dropna()
df
```

### Handling outliers

First, I will visualize the boxplots of the numerical variables to get the first sight of which columns may contain outliers. 

```python
#function for drawing boxplots of numerical variables:
def draw_numerical_boxplots(col_names):
    # Create subplots for multiple boxplots
    num_plots = len(numerical_vars_except_id)
    num_rows = 2
    num_cols = -(-num_plots // num_rows)  # Ceiling division to determine number of columns

    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))

    # Generate boxplots for each numerical variable
    for i, var in enumerate(col_names):
        row = i // num_cols
        col = i % num_cols
        sns.boxplot(data=df[var], ax=axes[row, col])
        axes[row, col].set_title(f"Boxplot of {var}")

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

numerical_vars = df.select_dtypes(include=['number']).columns.tolist()
numerical_vars_except_id = [col for col in numerical_vars if col != 'ID']

draw_numerical_boxplots(col_names = numerical_vars_except_id)
```

We can see some boxplots that have outliers laying outside of the whiskers. So the outliers appear in 7 variables: sessions, drives, total_sessions, total_navigations_fav1, total_navigations_fav2, driven_km_drives, duration_minutes_drives. Since there seem to be many outliers, I think we must keep the outliers since the results of our analysis are relied on the overall dataset.

### Categorical variables

In this section, I will check and transform to ensure the quality of the categorical variables.

Since the ID column is used for identify unique user, I will convert the column into categorical type to any change in calculation by mistake.

```python
df['ID'] = df['ID'].astype(object)
print(df['ID'].dtype)
```

The categorical variables in this dataset (except ID) are nominal variables. So I will check for the mistyped values in column label and device.

```python
# Get the list of categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Print the names of the categorical columns
print("Categorical columns:")
for col in categorical_cols:
    unique_values = df[col].unique()
    print(f"Unique values in column '{col}': {unique_values}")
```

So there are no mistyped values in these nominal variables. Now we can save the data for the analysis phase.

```python
df.to_csv('data/waze_dataset_cleaned.csv', index=False)
```
